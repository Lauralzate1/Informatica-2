En este ejercicio, crearás una clase llamada `MatrixOperations` que contendrá métodos para realizar varias operaciones con matrices. Además, implementarás métodos para generar diferentes tipos de matrices, como un tablero de ajedrez y una matriz con una cruz en el centro.

Aquí están los requerimientos detallados:

1. **Creación de Matrices Aleatorias:**
   - Crea una clase llamada `MatrixOperations`.
   - Esta clase debe tener un constructor que acepte un argumento opcional `seed` para inicializar la semilla del generador de números aleatorios. Si no se proporciona `seed`, debe utilizar un valor predeterminado de 42.
   - Debe tener un método llamado `randomMatrices` que cree dos matrices aleatorias con dimensiones 3x3 y 5x2 respectivamente. Los valores de las matrices deben estar en el rango de 1 a 10.
   
2. **Operaciones entre Matrices:**
   - Implementa métodos para realizar las siguientes operaciones entre matrices:
     - Suma
     - Resta
     - Multiplicación
   - Asegúrate de manejar adecuadamente las operaciones entre matrices de diferentes dimensiones, si es necesario.

3. **Tablero de Ajedrez:**
   - Crea un método llamado `chessBoard` que genere un tablero de ajedrez con dimensiones de 8x8.
   - Utiliza el carácter '#' para representar las casillas blancas y '.' para las casillas negras. Las casillas deben alternar entre blanco y negro.

4. **Matriz con Cruz:**
   - Implementa un método llamado `matrizCruz` que tome como argumento el tamaño de la matriz y devuelva una matriz cuadrada con una cruz de unos en el centro y ceros en el resto de las posiciones.

5. **Matriz con Diagonal Aleatoria:**
   - Crea un método llamado `matrizDiagonal` que genere una matriz cuadrada con dimensiones nxn.
   - Los valores en la diagonal principal deben ser aleatorios en el rango de 1 a 10, y todas las demás posiciones deben ser cero.

¡Tu tarea es implementar la clase `MatrixOperations` con estos métodos! Asegúrate de probar cada método para verificar su correcto funcionamiento.


```python
class MatrixOperations:
    def __init__(self, seed=42):
        np.random.seed(seed)
    
    def randomMatrices(self):
        matrix1 = np.random.randint(1, 10, size=(3, 3))
        matrix2 = np.random.randint(1, 10, size=(5, 2))
        return matrix1, matrix2
    
    def sumaMatrices(self, matrix1, matrix2):
        return np.add(matrix1, matrix2) #La funcion add sirve para sumar dos elementos
    
    def restaMatrices(self, matrix1, matrix2):
        return np.subtract(matrix1, matrix2) #La funcion subtract sirve para restar dos elementos
    
    def multipMatrices(self, matrix1, matrix2):
        return np.dot(matrix1, matrix2) #La funcion dot es para realizar el producto punto
    
    def chessBoard(self):
        chess_board = np.tile([['#', '.'], ['.', '#']], (4, 4)) #Se hace una matriz 2x2 y tile la duplica
        return chess_board
    
    def matrizCruz(self, size):
        matrix = np.zeros((size, size), dtype=int) #Es para llenar de ceros la matriz
        center = size // 2 #Se calcula el indice del centro div el tamaño por 2
        matrix[center, :] = 1 #Utilizando la variable center se llena la cruz de 1
        matrix[:, center] = 1
        return matrix
    
    def matrizDiagonal(self, n):
        matrix = np.zeros((n, n), dtype=int) #Lo mismo que en matrizCruz, llena de 0 la matriz
        np.fill_diagonal(matrix, np.random.randint(1, 10, size=n)) #Es para llenar la matriz principal con los numeros random
        return matrix
```

```python
matrix_ops = MatrixOperations()

# 1. Creación de matrices aleatorias
matrix1, matrix2 = matrix_ops.randomMatrices()
print("Matriz 1:\n", matrix1)
print("Matriz 2:\n", matrix2)

# 2. Operaciones entre matrices
#print("\nSuma de matrices:\n", matrix_ops.sumaMatrices(matrix1, matrix2))
#print("\nResta de matrices:\n", matrix_ops.restaMatrices(matrix1, matrix2))
#print("\nMultiplicación de matrices:\n", matrix_ops.multipMatrices(matrix1, matrix2))

# 3. Tablero de ajedrez
print("Tablero de ajedrez:")
print(matrix_ops.chessBoard())

# 4. Matriz con cruz
print("Matriz con cruz:")
print(matrix_ops.matrizCruz(5))

# 5. Matriz con diagonal aleatoria
print("Matriz con diagonal aleatoria:")
print(matrix_ops.matrizDiagonal(4))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def grafico_barras_notas(notas, color):
    asignaturas = list(notas.keys()) #Eje x
    valores = list(notas.values()) #Eje y
    fig, ax = plt.subplots()
    ax.bar(asignaturas, valores, color=color_barras, width= 0.2) #width es el grosor/ plt.bar es el grafico de barras
    plt.title('Notas por asignatura')
    plt.xlabel('Asignaturas')
    plt.ylabel('Notas')
    plt.show()

def grafico_dispersión_puntuaciones(puntuaciones_df, color, tamaño_puntos):
    plt.scatter(puntuaciones_df['Matemáticas'], puntuaciones_df['Ciencias'], color=color, s=tamaño_puntos) #plt.scatter es la dispersion
    plt.title('Relación entre puntuaciones de Matemáticas y Ciencias')
    plt.xlabel('Puntuaciones de Matemáticas')
    plt.ylabel('Puntuaciones de Ciencias')
    plt.show()
```

```python
# 1. Gráfico de barras de notas por asignatura
notas = {'Matemáticas': 85, 'Ciencias': 78, 'Historia': 92, 'Literatura': 88}
color_barras = ['gold', 'seagreen', 'pink', 'red']
grafico_barras_notas(notas, color_barras)

# 2. Gráfico de dispersión de puntuaciones de Matemáticas y Ciencias
data = {'Estudiante': ['A', 'B', 'C', 'D', 'E'],
        'Matemáticas': [75, 82, 90, 65, 88],
        'Ciencias': [70, 75, 85, 60, 92]}
puntuaciones_df = pd.DataFrame(data)
color_puntos = 'deepskyblue'
tamaño_puntos = 100
grafico_dispersión_puntuaciones(puntuaciones_df, color_puntos, tamaño_puntos)

```

```python
def estadisticas_ventas(ventas_df):
    total_ventas = ventas_df['Ventas'].sum()
    media_ventas = ventas_df['Ventas'].mean()
    desviacion_estandar = ventas_df['Ventas'].std()
    mes_max_ventas = ventas_df['Ventas'].idxmax()
    
    return pd.Series({'Total de ventas': total_ventas,
                      'Media mensual': media_ventas,
                      'Desviación estándar': desviacion_estandar,
                      'Mes con más ventas': mes_max_ventas})

def procesar_MMSE(df):
    # Ordenar el DataFrame por edad
    df_sorted = df.sort_values(by='Edad')
    
    # Contar el número de celdas que no son NaN para cada columna
    count_no_nan = df.count()
    
    # Invertir el DataFrame
    df_invertido = df_sorted.T
    
    # Organizar las columnas en orden alfabético
    df_ordenado = df_invertido.sort_index(axis=1)
    
    return df_ordenado

def plot_line_diagram(df, subject_id):
    # Extraer las pruebas de MMSE del sujeto
    pruebas = df.loc[subject_id][:-1]  # Excluir la columna 'Total'
    
    # Configurar el estilo del gráfico de líneas
    plt.plot(pruebas.index, pruebas.values, linestyle='--', color='red', marker='o', linewidth=3, label=subject_id)
    plt.legend(loc='lower right')
    plt.xlabel('Prueba')
    plt.ylabel('Resultado')
    plt.show()

def plot_pie_chart(df):
    total_mmse = df['Total']
    plt.pie(total_mmse, labels=total_mmse.index, autopct='%1.1f%%')
    plt.title('Total MMSE')
    plt.show()
```

```python
# 1. Estadísticas de ventas mensuales
ventas_data = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril'],
               'Ventas': [10000, 15000, 12000, 18000]}
ventas_df = pd.DataFrame(ventas_data)
estadisticas = estadisticas_ventas(ventas_df)
print("Estadísticas de ventas mensuales:\n", estadisticas)

# 2. Procesamiento del DataFrame MMSE
mmse_df = pd.read_csv('MMSE.csv')
mmse_procesado = procesar_MMSE(mmse_df)

# 3. Diagrama de líneas para pruebas MMSE de un sujeto
subject_id = 0  # Por ejemplo, seleccionamos el sujeto 0
plot_line_diagram(mmse_procesado, subject_id)

# 4. Gráfico circular para el total de MMSE
plot_pie_chart(mmse_procesado)
```

```python
import scipy.io as sio
```

```python
class analisisSenal:
    def __init__(self, archivo):
        self.data = sio.loadmat(archivo)
        self.senal1 = self.data['senal1']
        self.senal2 = self.data['senal2']
        self.cantidad_sensores = self.senal1.shape[1]
        self.num_puntos_epoca = self.senal1.shape[0]

    def valor_maximo(self, sensor):
        max_valor = np.max(self.senal1[:, sensor])
        max_posicion = np.argmax(self.senal1[:, sensor])
        return max_valor, max_posicion

    def graficar_senales(self, sensor):
        max_valor, max_posicion = self.valor_maximo(sensor)
        
        fig, ax = plt.subplots(2, 1, figsize=(10, 8))

        # Gráfico de la señal 1
        ax[0].plot(self.senal1[:, sensor], label='Señal 1', color='blue')
        ax[0].annotate(f'Máximo: {max_valor}', xy=(max_posicion, max_valor), xytext=(max_posicion, max_valor + 50),
                    arrowprops=dict(facecolor='red', shrink=0.05))
        ax[0].set_title('Señales')
        ax[0].set_xlabel('Tiempo')
        ax[0].set_ylabel('Amplitud')
        ax[0].legend()

        # Gráfico de la señal 2
        ax[1].plot(self.senal2[:, sensor], label='Señal 2', color='green')
        ax[1].annotate(f'Máximo: {max_valor}', xy=(max_posicion, max_valor), xytext=(max_posicion, max_valor + 50),
                    arrowprops=dict(facecolor='red', shrink=0.05))
        ax[1].set_xlabel('Tiempo')
        ax[1].set_ylabel('Amplitud')
        ax[1].legend()

        plt.tight_layout()
        plt.show()

    def graficar_histogramas(self):
        fig, ax = plt.subplots(2, 1, figsize=(10, 8))

        # Histograma de la señal 1
        ax[0].hist(self.senal1.flatten(), bins=50, color='blue', alpha=0.7, label='Señal 1')
        ax[0].axvline(np.mean(self.senal1), color='red', linestyle='--', label='Promedio')
        ax[0].axvline(np.median(self.senal1), color='green', linestyle='--', label='Mediana')
        ax[0].set_title('Histogramas de señales')
        ax[0].set_xlabel('Valor')
        ax[0].set_ylabel('Frecuencia')
        ax[0].legend()

        # Histograma de la señal 2
        ax[1].hist(self.senal2.flatten(), bins=50, color='green', alpha=0.7, label='Señal 2')
        ax[1].axvline(np.mean(self.senal2), color='red', linestyle='--', label='Promedio')
        ax[1].axvline(np.median(self.senal2), color='blue', linestyle='--', label='Mediana')
        ax[1].set_xlabel('Valor')
        ax[1].set_ylabel('Frecuencia')
        ax[1].legend()

        plt.tight_layout()
        plt.show()
```

```python
# Ejemplo
analisis = analisisSenal("Unidad 2\Taller\senales_potencial.mat")
print("Cantidad de sensores:", analisis.cantidad_sensores)
print("Número de puntos de época:", analisis.num_puntos_epoca)

# Graficar señales y histogramas
analisis.graficar_senales(sensor=0)
analisis.graficar_histogramas()
```
